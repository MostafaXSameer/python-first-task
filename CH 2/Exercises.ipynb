```python
import networkx as nx

G = nx.read_graphml('openflights_usa.graphml')

nx.draw(G, with_labels=True)
```

```python
print(G.nodes['FAI']) 
print(G.nodes['IND'])
nx.has_path(G, 'IND', 'FAI')  # true
```

```python
nx.shortest_path(G ,'IND', 'FAI')  # ['IND', 'BOS', 'SEA', 'FAI']
```

```python
nx.is_connected(G)  # false
```
